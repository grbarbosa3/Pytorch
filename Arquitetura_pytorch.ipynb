## nn.Sequential

O mdulo nn.Sequential  um container onde se pode colocar mltiplos mdulos. Ao realizar um forward em um objeto Sequential ele aplicar sequencialmente os mdulos nele contidas para gerar uma sada.

Segue abaixo um exemplo desse mdulo contendo 2 camadas Linear intercaladas por uma funo de ativao ReLU.


```python
import torch
from torch import nn # neural networks

input_size = 2
hidden_size = 8
output_size = 1

net = nn.Sequential(nn.Linear(in_features=input_size, out_features=hidden_size), # hidden (escondida)
                    nn.ReLU(),   # ativao no linear
                    nn.Linear(in_features=hidden_size, out_features=output_size)) # output (sada)
print(net)
```

O mdulo summary da biblioteca torchsummary nos permite visualizar mais informaes sobre a nossa rede, como quantidade de parmetros e o tamanho que cada elemento ocupa na memria.

Para mais informaes, acesse: https://github.com/sksq96/pytorch-summary

```python
from torchsummary import summary
summary(net, input_size=(1, input_size))
```

```python
print(X1.shape)
tensor = torch.from_numpy(X1).float()
pred   = net(tensor)
print(pred.size())
```

## Classe nn.Module

A forma mais organizada de definir modelos em PyTorch  implementando uma classe nos moldes da classe nn.Module. Para redes pequenas, como as que estamos aprendendo at o momento, sua importncia pode no se destacar, mas modelos maiores e com funcionalidades mais complexas, so mais fceis de implementar e realizar manutenes dessa forma.

Funes obrigatrias do nn.Module.



*   __init()__: definio de hiperparmetros e instncia do modelo
*   forward(): Fluxo da entrada para produzir uma sada




```python
class MinhaRede(nn.Module):
  
  def __init__(self, input_size, hidden_size, output_size):
    super(MinhaRede, self).__init__()
    
    # Definir a arquitetura
    self.hidden = nn.Linear(input_size, hidden_size)
    self.relu   = nn.ReLU()
    self.output = nn.Linear(hidden_size, output_size)
    
  def forward(self, X):
    
    # Gerar uma sada a partir do X
    hidden = self.relu(self.hidden(X))
    output = self.output(hidden)
    
    return output
```

Instanciando a rede

```python
input_size = 2
hidden_size = 8
output_size = 1

net = MinhaRede(input_size,hidden_size, output_size)  ## O mtodo __init__()
print(net)
```

## Forward

```python
print(X1.shape)
tensor = torch.from_numpy(X1).float()
pred   = net(tensor)
print(pred.size())
```

## Subindo informaes na GPU

Para conseguir executar modelos maiores em tempo hbil,  preciso carregar as informao na GPU para que o processamento seja realizado por ela. Dentre as informaes que conhecemos at o momento, duas delas precisam ser carregadas ba GPU:



*   Entrada
*   Rede


A clula a seguir contm a implementao sugerida para realizar esse processo.
Se certifique que o notebook est na configurao desejada (CPU/GPU). Para isso, basta acessar Edit -> Notebook Settings e definir o hardware desejado.

```python
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
  
print(device)
```

```python
input_size = 2
hidden_size = 8
output_size = 1

net = MinhaRede(input_size,hidden_size, output_size)  ## O mtodo __init__()
net = net.to(device)
print(net)
```

```python
print(X1.shape)
tensor = torch.from_numpy(X1).float()
tensor = tensor.to(device)
pred   = net(tensor)
print(pred.size())
```
